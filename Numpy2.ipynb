##<strong>선 색상과 스타일<strong>






*   plt.plot(x,y,color='red') -> red, greed, blue등 색상 이름
*   plt.plot(x,y,color'r') -> r,g,b,y,m,k등 색상 이름

*   plt.plot(x,y,color='0.2') -> 회색조 (0~1사이 값)
*   plt.plot(x,y,color='#ff0000') -> 16진수 색상 값

*   plt.plot(x,y,linestlye='solid') -> 실선('-')
*   plt.plot(x,y,linestlye='dashed') -> 파선('--')
*   plt.plot(x,y,linestlye='dashdot') -> 1점쇄선('-.')
*   plt.plot(x,y,linestlye='dotted') -> 점선(':')

```python
import plotly.express as px

x_ = np.array([1,2,3,4,5])
y_= x_ ** 2

fig = px.line(x=x_,y=y_)
fig.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go

korea_life = px.data.gapminder().query("country == 'Korea, Rep.'")
fig = px.line(korea_life,x='year',y='lifeExp',title='Life expectancy in Korea')
fig.show()
```

```python
korea_data = px.data.gapminder().query("country == 'Korea, Rep.'")
fig = px.scatter(korea_data, x='year', y='gdpPercap', title='한국인 GDP')
fig.show()
```
